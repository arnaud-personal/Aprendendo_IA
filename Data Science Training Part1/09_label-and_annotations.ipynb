```python
import numpy as np
from numpy.random import rand
import pandas as pd
from pandas import DataFrame, Series

import matplotlib.pyplot as plt
from matplotlib import rcParams

rcParams['figure.figsize']=5,4
```

```python
x=range(1,10)
y=[1,2,3,4,.5,4,3,2,1]
plt.bar(x,y)
plt.xlabel('Eixo X', loc='right')
plt.ylabel('Eixo Y', loc='top', labelpad=10)
```

```python
z = [1,2,3,4,.5]
plt.pie(z, 
        labels=['A','B','C','D','E'], 
        explode=[0,0,0,0.1,0],
        shadow=True,
        startangle=45
       )
plt.show()
```

```python
cars=pd.read_csv('mtcars.csv')
cars = cars.rename(columns={'Unnamed: 0':'car_name'})
mpg=cars['mpg']
fig=plt.figure(figsize=(8,4))
ax=fig.add_axes([.1,.1,1,1])
cars['mpg'].plot(kind='bar')
ax.set_xlabel('Car Names')
ax.set_ylabel("Miles/Galon")
ax.set_title("Miles per Galon", fontsize=20)
ax.set_xticklabels(cars['car_name'],rotation=90, fontsize='medium')
```

```python
# LEGENDAS
```

```python
plt.pie(z)
plt.legend(['A','B','C','D','E'], loc='right', bbox_to_anchor=(1.15, 0.5))
plt.show()
```

```python
fig=plt.figure(figsize=(8,4))
ax=fig.add_axes([.1,.1,1,1])

# Número de categorias
n = len(cars)
# Posições no eixo x
x = np.arange(n)

# Largura de cada barra
largura = 0.35
# Barras lado a lado
ax.bar(x - largura/2, cars['mpg'], width=largura, label='mpg')
ax.bar(x + largura/2, cars['qsec'], width=largura, label='qsec')

# Eixo X com os nomes
ax.set_xticks(x)
ax.set_xticklabels(cars['car_name'],rotation=90, fontsize='medium')

ax.set_xlabel('Car Names')
ax.set_ylabel("Miles/Galon")
ax.set_title("Miles per Galon", fontsize=20)

ax.legend()

plt.show()
```

```python
# ANNOTATIONS
```

```python
fig=plt.figure(figsize=(8,4))
ax=fig.add_axes([.1,.1,1,1])

# Número de categorias
n = len(cars)
# Posições no eixo x
x = np.arange(n)

# Barras lado a lado
ax.plot(cars['mpg'], label='MPG')
ax.plot(cars['qsec'], label='QSEC')

# Eixo X com os nomes
ax.set_xticks(x)
ax.set_xticklabels(cars['car_name'],rotation=90, fontsize='medium')

ax.set_xlabel('Car Names')
ax.set_ylabel("Miles/Galon")
ax.set_title("Miles per Galon", fontsize=20)

ax.legend()
ax.grid()

ax.set_ylim([0,45])
ax.annotate(
    'Toyota Corolla', 
    fontsize=12,
    color='red',
    xy = (19, 33.9), 
    xytext=(12,35), 
    arrowprops=dict(arrowstyle='fancy', color='red')
)

```
