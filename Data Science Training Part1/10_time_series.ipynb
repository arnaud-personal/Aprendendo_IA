```python
import numpy as np
from numpy.random import rand
import pandas as pd
from pandas import DataFrame, Series
import matplotlib.pyplot as plt
```

```python
import chardet
with open('Superstore-Sales.csv', 'rb') as f:
    result = chardet.detect(f.read(10000))
    print(result)
```

```python
df = pd.read_csv(
    'Superstore-Sales.csv', 
    encoding='Windows-1252', 
    index_col='Order Date', 
    parse_dates=True
)
df=df.sort_index()
df.head()
```

```python
# TIME SERIES COM FUNÇÕES
#=========================

df['Order Quantity'].plot()
```

```python
# Usando um sample do original
df2 = df.sample(n=100, random_state=25, axis=0)
df2 = df2.sort_index()
plt.xlabel('Order Date')
plt.ylabel('Order Quantity')
plt.title('Superstore Sales')
df2['Order Quantity'].plot()
```

```python
# TIME SERIES COM OBJETO
#=========================
```

```python
fig, ax = plt.subplots(figsize=(10,5))
ax.set_title('Superstore Sales')
ax.set_xlabel('Order Date')
ax.set_ylabel('Order Quantity')
ax.plot(df2['Order Quantity'])
```
