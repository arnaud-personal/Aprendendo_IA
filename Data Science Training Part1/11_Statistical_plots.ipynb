```python
import numpy as np
from numpy.random import rand
import pandas as pd
from pandas import DataFrame, Series
import matplotlib.pyplot as plt
import seaborn as sb
sb.set_style('whitegrid')
```

```python
df = pd.read_csv('mtcars.csv')
df=df.rename(columns={'Unnamed: 0':'car name'})
df=df.set_index('car name')
df.head()
```

```python
# =========================
#       HISTOGRAMAS
# =========================
```

```python
# MATPLOTLIB FUNCTIONS
#---------------------

df['mpg'].plot(kind='hist')
```

```python
plt.hist(df['mpg'])
plt.plot()
```

```python
# MATPLOTLIB OBJECTS
#-------------------

fig, ax = plt.subplots()
ax.hist(df['mpg'])
```

```python
# SEABORN FUNCTIONS
#---------------------

sb.histplot(
    df['mpg'], 
    kde=True, 
    kde_kws=dict(cut=3), 
    bins=12
)
sb.rugplot(df, x="mpg")
```

```python
# =========================
#       SCATTER PLOTS
# =========================
```

```python
# MATPLOTLIB FUNCTIONS
#---------------------

df.plot(kind='scatter', x='hp', y='mpg', s=75)
```

```python
# SEABORN FUNCTIONS
#---------------------

sb.scatterplot(
    df, 
    x='hp', 
    y='mpg', 
    size='disp'
)
```

```python
sb.regplot(
    data=df, 
    x='hp', 
    y='mpg', 
    scatter=True
)
```

```python
# =========================
#   SCATTER PLOTS MATRIX
# =========================
```

```python
sb.pairplot(df, vars=['mpg','disp','hp','wt'],kind="reg")
```

```python
# =========================
#         BOXPLOTS
# =========================
```

```python
# MATPLOTLIB
#------------
df.boxplot(column='mpg',by='am')
```

```python
df.boxplot(column='mpg',by='am')
df.boxplot(column='wt',by='am')
```
