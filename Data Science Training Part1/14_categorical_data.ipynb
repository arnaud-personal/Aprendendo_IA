```python
cars = pd.read_csv('mtcars.csv')
cars = cars.rename(columns={'Unnamed: 0':'car name'})
cars = cars.set_index('car name')
cars.head(3)
```

```python
cars.carb.value_counts()
```

```python
cars_cat = cars[['cyl','vs','am','gear','carb']]
cars_cat.head(3)
```

```python
gear_groups = cars_cat.groupby('gear')
gear_groups.describe()
```

```python
gear_groups.mean()
```

```python
cars_cat.mean()
```

```python

#--------------------------------------------------------
# TRANSFORMANDO VARIÁVEL EM TIPO DE DATA DATA CATEGÓRICO
#--------------------------------------------------------
```

```python
cars['group'] = pd.Series(cars.gear, dtype='category')
cars.head()
```

```python
cars.group.dtype
```

```python
cars.group.value_counts()
```
