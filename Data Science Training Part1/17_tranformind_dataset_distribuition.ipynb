```python
import numpy as np
import pandas as pd

import matplotlib as plt
import seaborn as sb
from pylab import rcParams

import sklearn
from sklearn import preprocessing
from sklearn.preprocessing import scale
```

```python
rcParams['figure.figsize']=[8,4]
plt.style.use('seaborn-v0_8-whitegrid')
```

```python
cars = pd.read_csv('mtcars.csv')
cars = cars.rename(columns={'Unnamed: 0':'car_name'})
cars = cars.set_index('car_name')
cars.head()
```

```python
cars.mpg.plot()
```

```python
cars.mpg.describe()
```

```python
# SCALE
#----------------------------------
```

```python
mpg_matrix = cars.mpg.values.reshape(-1,1)
scaled = preprocessing.MinMaxScaler().fit_transform(mpg_matrix)
plt.pyplot.plot(scaled)
```

```python
scaled = preprocessing.MinMaxScaler(feature_range=(0,10)).fit_transform(mpg_matrix)
plt.pyplot.plot(scaled)
```

```python

# STANDARDIZATION
#----------------------------------
```

```python
std_mpg = scale(
    cars.mpg,
    axis=0, 
    with_mean=False, 
    with_std=False 
)
plt.pyplot.plot(std_mpg)
```

```python
std_mpg = scale(cars.mpg)
plt.pyplot.plot(std_mpg)
```
