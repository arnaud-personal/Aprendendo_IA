```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sb
from pylab import rcParams

rcParams['figure.figsize']=[8,4]
plt.style.use('seaborn-v0_8-whitegrid')
```

```python
df = pd.read_csv('iris.data.csv',header=None)
df.columns=['Sepal Length','Sepal Width','Petal Length','Petal Width','Species']
df.head()
```

```python
sb.boxplot(data=df, x='Species', y='Sepal Length', palette='hls', hue='Species')
```

```python
sb.pairplot(df, hue='Species', palette='hls')
```

```python

first_quantil = df[df['Species'] == 'virginica']['Sepal Length'].quantile(0.25)
third_quantil = df[df['Species'] == 'virginica']['Sepal Length'].quantile(0.75)
IQR = third_quantil - first_quantil
f=df[df['Species'] == 'virginica']['Sepal Length'] < (first_quantil - 1.5*IQR)
indices_true = f[f].index.tolist()
indices_true
```

```python
df.iloc[indices_true,0]
```
