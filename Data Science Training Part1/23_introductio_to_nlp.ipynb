```python
import nltk
```

```python
text = "On Wednesday, the Association for Computing Machinery, the world’s largest society of computing professionals, announced that Hinton, LeCun and Bengio had won this year’s Turing Award for their work on neural networks. The Turing Award, which was introduced in 1966, is often called the Nobel Prize of computing, and it includes a $1 million prize, which the three scientists will share."
```

```python
nltk.download('punkt')
nltk.download('punkt_tab')
```

```python

# ---> SENTENCE TOKENIZER

from nltk.tokenize import sent_tokenize
sent_tk = sent_tokenize(text)
print(type(sent_tk))
print('Sentece tokenized text:') 
for token in sent_tk:
    print('--->', token)
```

```python

# ---> WORD TOKENIZER

from nltk.tokenize import word_tokenize
word_tk = word_tokenize(text)
print(type(word_tk))
print('Word tokenized text:') 
print(word_tk)
```

```python

# ---> REMOVING STOP WORDS

from nltk.corpus import stopwords
nltk.download('stopwords')
sw = set(stopwords.words('english'))
print('Stop Word in English language:') 
print(sw)
```

```python
filtered_words = [w for w in word_tk if w not in sw]
print('Text without stop words:') 
print(filtered_words)
```

```python
import string
clean_tokens = [palavra for palavra in filtered_words if palavra not in string.punctuation]
print('Text without stop words and pontuation:') 
print(clean_tokens)
```

```python

# ---> STEMMING

from nltk.stem import PorterStemmer
port_stem = PorterStemmer()
```

```python
stemmed_words = [port_stem.stem(w) for w in clean_tokens]
print('Stemmed Text:') 
print(stemmed_words)
```

```python

# ---> LEMMATIZING

nltk.download('wordnet')
```

```python
from nltk.stem.wordnet import WordNetLemmatizer
lem =  WordNetLemmatizer()
lem_words = [lem.lemmatize(w) for w in clean_tokens]
print('Lemmatized words:') 
print(lem_words)
```

```python

# ---> PARTS OF SPEECH TAGGING

nltk.download('averaged_perceptron_tagger')
nltk.download('averaged_perceptron_tagger_eng')
```

```python
from nltk import pos_tag
pos_tagged_words = pos_tag(word_tk)
print('Tagged words:') 
print(pos_tagged_words)
```

```python

# ---> FREQUENCY DISTRIBUTION PLOTS

from nltk.probability import FreqDist
print('WORD TOKENIZED:\n\n',word_tk)
fd = FreqDist(word_tk)
print("\nf:", fd)
```

```python
import matplotlib.pyplot as plt
fd.plot(30, cumulative=False)
plt.show()
```

```python
fd = FreqDist(clean_tokens)
fd.plot(30, cumulative=False)
plt.show()
```

```python
fd_alpha = FreqDist(text)
fd_alpha.plot(30, cumulative=False)
plt.show()
```
