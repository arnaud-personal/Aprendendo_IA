```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from pylab import rcParams

import sklearn
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import scale
```

```python
rcParams['figure.figsize'] = 10, 8
```

<h2>Criando dados fict√≠cios</h2>

```python
rooms = 2*np.random.rand(100,1)+3
price = 265 + 6*rooms + abs(np.random.randn(100,1))
```

<h2>Criando o Plot</h2>

```python
plt.plot(rooms, price, 'o')
plt.xlabel('# of rooms')
plt.ylabel('price')
```

<h2>Criando e treinando a regress√£o linear</h2>

```python
modelo = LinearRegression()

modelo.fit(rooms, price)
```

<h2>Obtendo par√¢metros e fazendo predi√ß√µes</h2>

<h3>Coeficiente angular e intercepto</h3>

```python
# Coeficiente angular (a)
print('Coeficiente (a):', modelo.coef_[0])

# Intercepto (b)
print('Intercepto (b):', modelo.intercept_)
```

<h2>Desempenho do modelo</h2>

```python
print(f'Desempenho do modelo (R¬≤):  {modelo.score(rooms, price):.2f}')
```

<h2>Visualizando o resultado</h2>

<h3>Obtendo a previs√£o de pre√ßos segundo o modelo</h3>

```python
price_pred = modelo.predict(rooms)
```

<h3>Plot (matplotlib)</h3>

```python
# Plotando os pontos originais
plt.scatter(rooms, price, color='blue', marker='o' , label='Dados reais')

# Plotando a reta de regress√£o
plt.plot(rooms, price_pred, color='red', label='Regress√£o Linear')

plt.title('Regress√£o Linear Simples')
plt.xlabel('# of rooms')
plt.ylabel('price')
plt.legend()
plt.grid(True)
plt.show()
```

<h3>Fazendo predi√ß√£o de um √∫nico valor</h3>

```python
x = int(input('# de ambientes: '))
pred = modelo.predict([[x]])
print(f'Valor para {x} ambientes: {pred[0][0]:.2f}')
```
